```python
# Часть 1 №1 функция returnSubImage
def returnSubImage(imageMatrix, startingRow, startingCol, sizeOfSquare = 100):
    subImageMatrix = []
    for idx in range(sizeOfSquare):
        subImageMatrix.append([])
        for idy in range(sizeOfSquare):
            subImageMatrix[idx].append(imageMatrix[idx][idy].copy())
            
    if (np.size(imageMatrix, 0) - sizeOfSquare - startingRow) < 0:
        return "Ошибка, вырезаемое изображение по x больше исходного"
    if (np.size(imageMatrix, 1) - sizeOfSquare - startingCol) < 0:
        return "Ошибка, вырезаемое изображение по y больше исходного"
    
    for idx in range(startingRow, startingRow + sizeOfSquare):
        for idy in range(startingCol, startingCol + sizeOfSquare):
            subImageMatrix[idx - startingRow][idy - startingCol] = imageMatrix[idx][idy]
    return subImageMatrix
```

```python
# Часть 1 №2 вывод части изображения с началом координат 100х100
img2 = returnSubImage(img, 200, 200)
```

```python
# Часть 1 №3 вывод части изображения с началом координат 150х200
img3 = returnSubImage(img, 150, 200)
```

```python
imgplot = plt.imshow(img2)
#Вывод изображения при помощи команды imshow()
```

```python
imgplot = plt.imshow(img3)
#Вывод изображения при помощи команды imshow()
```
