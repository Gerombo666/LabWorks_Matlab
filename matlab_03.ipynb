```python
def imadjust(x,a,b,c,d,gamma=1):
    # Similar to imadjust in MATLAB.
    # Converts an image range from [a,b] to [c,d].
    # The Equation of a line can be used for this transformation:
    #   y=((d-c)/(b-a))*(x-a)+c
    # However, it is better to use a more generalized equation:
    #   y=((x-a)/(b-a))^gamma*(d-c)+c
    # If gamma is equal to 1, then the line equation is used.
    # When gamma is not equal to 1, then the transformation is not linear.

    y = (((x - a) / (b - a)) ** gamma) * (d - c) + c
    return y
```

```python
# Часть 1
# Растяжение контраста и демонстрация при Е = 1..6
float_epsilon = np.finfo(float).eps
img4 = mpimg.imread('./matlab/wheel.png') #форматы изображения: png, jpg...
m = np.mean(img4)

contrast1 = 1.0 / (1 + (m / (img4 + float_epsilon))** 1)
contrast2 = 1.0 / (1 + (m / (img4 + float_epsilon))** 2)
contrast3 = 1.0 / (1 + (m / (img4 + float_epsilon))** 3)
contrast4 = 1.0 / (1 + (m / (img4 + float_epsilon))** 4)
contrast5 = 1.0 / (1 + (m / (img4 + float_epsilon))** 5)
contrast6 = 1.0 / (1 + (m / (img4 + float_epsilon))** 6)
```

```python
# Часть 3
img5 = mpimg.imread('./matlab/two_cats.png') #форматы изображения: png, jpg...
from scipy.ndimage import filters

im_original = img5.copy()

# Construct two ndarrays of same size as the input image
imx = np.zeros(im_original.shape)
imy = np.zeros(im_original.shape)

# Run the Sobel operator
filters.sobel(im_original,1,imx,cval=0.0)  # axis 1 is x
filters.sobel(im_original,0,imy, cval=0.0) # axis 0 is y

imx = np.clip((1 * imx + 0), 0, 1)
imy = np.clip((1 * imy + 0), 0, 1)

imx_imy = imx+imy
imx_imy = np.clip((1 * imx + 0), 0, 1)


# Construct the plot
fig = plt.figure(figsize=(17,14))

ax1 = fig.add_subplot(441)
ax1.set_title('Original')
ax1.axis('off')
ax1.imshow(im_original, cmap='gray')

ax2 = fig.add_subplot(442)
ax2.set_title('Вертикальные края')
ax2.axis('off')
ax2.imshow(imx, cmap='gray')

ax3 = fig.add_subplot(443)
ax3.set_title('Горизонтальные края')
ax3.axis('off')
ax3.imshow(imy, cmap='gray')

ax4 = fig.add_subplot(444)
ax4.set_title('Сложение H и L')
ax4.axis('off')
ax4.imshow(imx_imy, cmap='gray')
```
