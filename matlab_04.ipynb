```python
# Часть 2 №1
img = mpimg.imread('./matlab/building.png') #форматы изображения: png, jpg...
img2 = mpimg.imread('./matlab/text.png') #форматы изображения: png, jpg...
liftingbody_r0 = mpimg.imread('./matlab/liftingbody_r0.png') #форматы изображения: png, jpg...
#img = mpimg.imread('./matlab/img6.jpg') #форматы изображения: png, jpg...
#Чтение изображения и сохранение в переменную img в виде матрицы
print(img2)
```

```python
# Часть 1
from skimage import data,filters,feature

#img = data.camera()
edges1 = feature.canny(img)   #sigma=1
edges2 = feature.canny(img,sigma=1.4)   #sigma=3
#edges2 = filters.prewitt(img)
#edges2 = filters.scharr(img)
edges3 = filters.roberts(img)
#edges3 = filters.sobel(img)

# Construct the plot
fig = plt.figure(figsize=(17,14))

ax1 = fig.add_subplot(441)
ax1.set_title('Original')
ax1.axis('off')
ax1.imshow(img, cmap='gray')

ax2 = fig.add_subplot(442)
ax2.set_title('canny sigma = 1')
ax2.axis('off')
ax2.imshow(edges1, cmap='gray')

ax3 = fig.add_subplot(443)
ax3.set_title('canny sigma = 1.4')
ax3.axis('off')
ax3.imshow(edges2, cmap='gray')

ax4 = fig.add_subplot(444)
ax4.set_title('roberts')
ax4.axis('off')
ax4.imshow(edges3, cmap='gray')
```
