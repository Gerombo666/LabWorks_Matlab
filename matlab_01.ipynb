```python
# Часть 1 №1 создайте 3 матрицы
A = np.array([[22, 5],
              [4, 11],
              [13, 2]])

B = np.array([[17, 9, 2, 1, 3],
              [6, 4, 11, 9, 16]])

C = np.array([[1, 0, 0],
              [0, 1, 0],
              [0, 0, 1]])
```

```python
# Часть 1 №2 вычитание матриц
D = A-np.array([[0, 0],
                [3, 0],
                [0, 0]])
print(D)
```

```python
# Часть 1 №2 умножение матриц
E = A*3

print(E)
```

```python
# Часть 1 №2 транспонирование матриц
F = np.transpose(A)

print(F)
```

```python
# Часть 1 №3 сложение матриц
G = B + np.array([[0, 0, 1, 0, 0],
                  [0, 0, -8, 0, 0]])

print(G)
```

```python
# Часть 1 №3 транспонирование матриц
print(G, "\n")

H = G[:,0::2]
#Выбираем каждый 2 элемент

print(H, "\n")

H = H[:,::-1]
#Переворачиваем матрицу

print(H)
```

```python
# Часть 1 №4
I = C[0:2,0:2]

print(I)
```

```python
# Часть 1 №4
J = C[0:2,0:2] * np.array([[1, 1],[1, -1]])

print(J)
```

```python
# Часть 1 №4
K = A * np.array([[1, -1],[1, -1],[1, -1]])

print(K)
```

```python
#Нахождение минимума и максимума в матрице
min = np.amin(img)
max = np.amax(img)

print(min)
print(max)
```

```python
# Копируем изображение в img5
img5 = img.copy()

# Находим min и max
min = np.amin(img)
max = np.amax(img)

for idx, val in enumerate(img):
    for jdx, val2 in enumerate(val):
        img5[idx][jdx] = (img[idx][jdx]-min)/(max-min)
```

```python
#Минимальная и максимальная яркость изображения до обработки
print(np.amin(img))
print(np.amax(img))
```

```python
#Минимальная и максимальная яркость изображения после обработки
print(np.amin(img5))
print(np.amax(img5))
```

```python
imgplot = plt.imshow(img5)
#вывод картинки светлее оригинала
```
